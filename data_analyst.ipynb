## Data Analyst

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('air_fryers_clean_brand_year.csv')
```

```python
print("Number of brands:", df['brand'].nunique())
print("Years covered:", df['year'].min(), "-", df['year'].max())
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.lineplot(data=df, x='year', y='avg_price', hue='brand', ax=axes[0], marker='o')
axes[0].set_title('Average Price over Time')
axes[0].set_xticks(df['year'].unique())

sns.lineplot(data=df, x='year', y='avg_rating', hue='brand', ax=axes[1], marker='o')
axes[1].set_title('Average Rating over Time')
axes[1].set_xticks(df['year'].unique())

sns.lineplot(data=df, x='year', y='brand_share', hue='brand', ax=axes[2], marker='o')
axes[2].set_title('Brand Market Share over Time')
axes[2].set_xticks(df['year'].unique())

plt.tight_layout()
plt.show()
```

```python
feature_cols =[
    'compact_share', 
    'dual_basket_share', 
    'oven_style_share', 
    'rotisserie_share', 
    'window_share'
]

print(df[feature_cols].mean().sort_values(ascending=False))
print("\n")
print(df.groupby('brand')[feature_cols].mean())
```

**Product Characteristics**
* **Common:** Compact design is nearly universal across all brands. Oven-style designs are also highly prevalent.
* **Rare:** Dual-basket and window features are extremely rare across the market.
* **Specialization:** Brands like Cuisinart, Dash, and Oster focus almost exclusively on oven-style fryers. Chefman and GoWISE USA stand out for incorporating rotisserie features, while Chefman is the only brand significantly investing in window features.

**Market Description**
* **Expensive Brands:** Cuisinart is by far the most expensive brand, consistently pricing above $200, followed by Oster and Ninja.
* **Market Share:** Early in the dataset (2019), GoWISE USA, Dash, and Ninja dominated. By 2022 and 2023, Instant Pot and Ninja captured the largest market shares, while GoWISE and Dash saw significant declines.
* **Stability:** The market is highly dynamic rather than stable. While average prices and ratings remain relatively flat over time for individual brands, market share fluctuates wildly as consumer preferences shift and dominant players rotate.
